```python
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
from sklearn import linear_model, preprocessing
import pandas as pd
import numpy as np
```

```python
#load the data
data = pd.read_csv("datasets/car.data")
data.head()
```

```python
# convert the non-nummerical data to numerical data
#use preprocessing. It uses lists

le = preprocessing.LabelEncoder()
buying = le.fit_transform(list(data["buying"]))                 #take the entire buying column and convert it
maintenance = le.fit_transform(list(data["maintenance"]))
door = le.fit_transform(list(data["door"]))
persons = le.fit_transform(list(data["persons"]))
lug_boot = le.fit_transform(list(data["lug_boot"]))
safety = le.fit_transform(list(data["safety"]))
car_class = le.fit_transform(list(data["class"]))
```

```python
predict = "class"

x = list(zip(buying, maintenance, door, persons, lug_boot, safety))     #features
y = list(car_class)         #labels

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size=0.1)
print(x_train, y_test)

```

```python
# define k 
model = KNeighborsClassifier(n_neighbors=7)
model.fit(x_train, y_train)

# determine the accuracy of our model
accuracy = model.score(x_test, y_test)
accuracy
```

```python
# test our model on the data
predicted = model.predict(x_test)

# classes for which the data is supposed to be
names = ["unacc", "acc", "good", "vgood"]

count = 0
for x in range(len(predicted)):
    print("Predicted class: ", names[predicted[x]],  "Data: ", x_test[x],  "Original class: ", names[y_test[x]])
    count += 1
    if count == 20:
        break
   
```
